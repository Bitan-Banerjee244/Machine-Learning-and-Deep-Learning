# Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = [
    "The cat sat on the mat.",
    "The dog sat on the log.",
    "Cats and dogs are common pets.",
    "The mat was sat on by the cat.",
    "Dogs bark loudly at strangers."
]
```

```python
vectorizer = CountVectorizer(ngram_range=(2,2)) # ngrams make groups among vocabulary
```

```python
X = vectorizer.fit_transform(documents)
```

```python
print("Vocalbulary",vectorizer.get_feature_names_out()) # All Vocabulary
print("\nBoW Matrix\n",X.toarray())
```

# TF IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents = [
    "The cat sat on the mat.",
    "The dog sat on the log.",
    "Cats and dogs are common pets.",
    "The mat was sat on by the cat.",
    "Dogs bark loudly at strangers."
]
```

```python
vectorizer = TfidfVectorizer()
```

```python
X = vectorizer.fit_transform(documents)
```

```python
print("Vocalbulary",vectorizer.get_feature_names_out()) # All Vocabulary
print("\nBoW Matrix\n",X.toarray())
```
